- 1. 8-Puzzle Problem

```python
from collections import deque

# Function to check if the puzzle is solvable
def is_solvable(puzzle):
    inversions = 0
    flat_puzzle = [tile for row in puzzle for tile in row if tile != 0]
    for i in range(len(flat_puzzle)):
        for j in range(i + 1, len(flat_puzzle)):
            if flat_puzzle[i] > flat_puzzle[j]:
                inversions += 1
    return inversions % 2 == 0

# Function to find the position of the blank (0) tile
def find_blank(puzzle):
    for i in range(len(puzzle)):
        for j in range(len(puzzle[i])):
            if puzzle[i][j] == 0:
                return i, j

# BFS Implementation
def bfs_8_puzzle(start, goal):
    moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]  # Right, Left, Down, Up
    visited = set()
    queue = deque([(start, [])])
    while queue:
        current_state, path = queue.popleft()
        visited.add(tuple(tuple(row) for row in current_state))
        if current_state == goal:
            return path
        blank_row, blank_col = find_blank(current_state)
        for move in moves:
            new_row, new_col = blank_row + move[0], blank_col + move[1]
            if 0 <= new_row < 3 and 0 <= new_col < 3:
                new_state = [row[:] for row in current_state]
                new_state[blank_row][blank_col], new_state[new_row][new_col] = (
                    new_state[new_row][new_col],
                    new_state[blank_row][blank_col],
                )
                if tuple(tuple(row) for row in new_state) not in visited:
                    queue.append((new_state, path + [new_state]))

# Start and Goal States
start_state = [
    [8, 6, 7],
    [2, 5, 4],
    [3, 0, 1],
]
goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0],
]

if is_solvable(start_state):
    solution = bfs_8_puzzle(start_state, goal_state)
    print("Solution Path:")
    for step in solution:
        for row in step:
            print(row)
        print()
else:
    print("The given puzzle is not solvable.")
```

# 2. 8-Queens Problem

```python
def is_safe(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, len(board)), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True

def solve_n_queens_dfs(board, col):
    if col >= len(board):
        return True
    for i in range(len(board)):
        if is_safe(board, i, col):
            board[i][col] = 1
            if solve_n_queens_dfs(board, col + 1):
                return True
            board[i][col] = 0
    return False

def print_board(board):
    for row in board:
        print(" ".join("Q" if cell == 1 else "." for cell in row))
    print()

# Solve the 8-Queens problem
n = 8
chess_board = [[0 for _ in range(n)] for _ in range(n)]
if solve_n_queens_dfs(chess_board, 0):
    print("One solution to the 8-Queens problem:")
    print_board(chess_board)
else:
    print("No solution exists.")
```
