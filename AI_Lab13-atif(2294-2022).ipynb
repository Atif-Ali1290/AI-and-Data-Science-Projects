```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load the dataset provided by the user
df  = pd.read_csv("Mall_Customers.csv")
# mall_data = pd.read_csv(dataset_path)

# Display the first few rows to understand its structure
df.head()
```

### Task 1: Distribution plot for the annual income

```python
# Task 1: Distribution plot for Annual Income
plt.figure(figsize=(10, 6))
sns.histplot(df["Annual Income (k$)"], kde=True, color='blue', bins=20)
plt.title("Distribution of Annual Income of Customers", fontsize=16)
plt.xlabel("Annual Income (k$)", fontsize=12)
plt.ylabel("Frequency",fontsize=12 )
plt.show()
```

### Task 2: Line plot for the relationship between:
- Annual Income vs. Age
- Annual Income vs. Spending Score

#### Task 2a: Line plot - Annual Income vs Age

```python

plt.figure(figsize=(10, 6))
sns.lineplot(x=df["Age"], y=df["Annual Income (k$)"], color='green', marker="o")
plt.title("Relationship Between Age and Annual Income", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Annual Income (k$)", fontsize=12)
plt.grid()
plt.show()


```

#### Task 2b: Line plot - Annual Income vs Spending Score

```python

plt.figure(figsize=(10, 6))
sns.lineplot(x=df["Annual Income (k$)"], y=df["Spending Score (1-100)"], color='purple', marker="o")
plt.title("Relationship Between Annual Income and Spending Score", fontsize=16)
plt.xlabel("Annual Income (k$)", fontsize=12)
plt.ylabel("Spending Score (1-100)", fontsize=12)
plt.grid()
plt.show()
```
